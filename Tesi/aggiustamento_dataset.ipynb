```python
import os
import glob
from pydub import AudioSegment

video_dir = '/content/drive/MyDrive/Tesi/my_dataset/Actor_07'  # Path where the videos are located
extension_list = ('*.m4a')

os.chdir(video_dir)
for extension in extension_list:
    for video in glob.glob(extension):
        wav_filename = os.path.splitext(os.path.basename(video))[0] + '.wav'
        print(wav_filename)
        AudioSegment.from_file(video).export(wav_filename, format='wav')
```

```python
import re
pattern = re.compile(".wav")

for extension in extension_list:
    for video in glob.glob(extension):
            filename_extension = os.path.splitext(os.path.basename(video))[1]
            if(re.search(pattern, filename_extension)):
                os.remove(os.path.splitext(os.path.basename(video))[0] + '.m4a')
```
